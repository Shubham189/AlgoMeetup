````python
def inorder_iterative(root):
    if root is None:
        return 
    
    stack = []
    while stack or root is not None:
        if root is not None:
            print("appending here", root.value)
            stack.append(root)
            root = root.left
        else:
            node = stack.pop()
            print(node.value)
            root = node.right
        
        
````

````python
inorder_iterative(root)
````

````python
def preorder_iterative(root):
    if root is None:
        return 
    
    stack = []
    while stack or root is not None:
        if root is not None:
            print(root.value)
            stack.append(root)
            root = root.left
        else:
            node = stack.pop()
            root = node.right
````

````python
preorder_iterative(root)
````

````python
def postorder_iterative(root):
    if root is None:
        return 
    
    stack = []
    while stack or root:
        if root is not None:
            stack.append(root)
            root = root.left
        else:
            node = stack[-1]
            if node.right is not None:
                root = node.right
            else:
                stack.pop()
                print(node.value)
                
                while stack and node is stack[-1].right:
                    node = stack.pop()
                    print(node.value)
            
        
````

````python
postorder_iterative(root)
````

````python
def size(root):
    if root is None:
        return 0
    return height(root.left) + height(root.right) + 1
````

````python
size(root)
````

````python
def height(root):
    if root is None:
        return 0
    return 1 + max(height(root.left), height(root.right))
````

````python
height(root)
````

````python
def level_order(root):
    if root is None:
        return
    q = []
    q.append(root)
    while q:
        node = q.pop(0)
        print(node.value)
        
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
````

````python
level_order(root)
````

````python
def lca(root, a, b):
    
    if root is None:
        return None
    if root is a or root is b:
        return root
    
    left = lca(root.left, a, b)
    right = lca(root.right, a, b)
    
    if (left is not None) and (right is not None):
        return root
    return left if left is not None else right
````

````python
def bfs(root, x):
    if root is None:
        return
    q = []
    q.append(root)
    while q:
        node = q.pop(0)
        print(node.value)
        if x == node.value:
            print(f'Found at {node.value}')
            return
        
        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)
````

````python
bfs(root, 5)
````

````python
def dfs(root, x):
    if root is None:
        return 
    print(root.value)
    if root.value == x:
        print(f'Found at {root.value}')
        return x
    left = dfs(root.left,x)
    dfs(root.right,x) if left is None else left
````

````python
dfs(root, 7)
````

## Binary Search Tree

````python
root = Node(4)

root.left = Node(2)
root.right = Node(6)

root.left.left = Node(1)
root.left.right = Node(3)

root.right.left = Node(5)
root.right.right = Node(7)

print(root)
````

````python
def is_bst(root):
    
    def bst_helper(root, minimum, maximum):
        if root is None:
            return True
        if root.value < minimum or root.value > maximum:
            return False
        
        return bst_helper(root.left, minimum, root.value) and bst_helper(root.right, root.value, maximum)
            
    return bst_helper(root, -float("inf"), float("inf"))
        
````

````python
is_bst(root)
````
